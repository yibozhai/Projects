```python
import numpy as np

# create a custom function
def calculate_profit(y, y_pred):
    tn = np.where((y_pred==0) & (y==0), 1500, 0)
    fn = np.where((y_pred==0) & (y==1), -5000, 0)
    return np.sum([tn,fn])


# add metric to PyCaret
add_metric('profit', 'Profit', calculate_profit)
```

### Now Let's look at if profit changes when we change the threshold

```python
from pycaret.classification import setup, add_metric, create_model, predict_model, pull

# Set up the classification environment
clf1 = setup(data=cc_data, target='DEFAULT', session_id=123)

# Add the custom profit metric
add_metric('profit', 'Profit', calculate_profit)
```

```python
import pandas as pd

# Function to evaluate profit at different thresholds
def evaluate_thresholds(model, data, target, thresholds):
    results = []
    for threshold in thresholds:
        # Get prediction probabilities
        predictions = predict_model(model, data=data, raw_score=True)
        y_true = data[target]
        y_proba = predictions['prediction_score_1']  # Assuming 'Score' is the predicted probability of the positive class
        
        # Apply threshold
        y_pred = (y_proba >= threshold).astype(int)
        
        # Calculate profit
        profit = calculate_profit(y_true, y_pred)
        results.append((threshold, profit))
    
    # Create a DataFrame with the results
    results_df = pd.DataFrame(results, columns=['Threshold', 'Profit'])
    return results_df

# Train the model
model = create_model('lr')

# Evaluate the profit at different thresholds
thresholds = np.arange(0.0, 1.05, 0.05)
results_df = evaluate_thresholds(model, cc_data, 'DEFAULT', thresholds)

# Print the results
print(results_df)

# Find the best threshold
best_threshold = results_df.loc[results_df['Profit'].idxmax()]
print(f"Best Threshold: {best_threshold['Threshold']}, Highest Profit: {best_threshold['Profit']}")
```

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Define the columns of interest for payment and bill amounts
pay_amt_columns = [f'PAY_AMT{i}' for i in range(1, 7)]
bill_amt_columns = [f'BILL_AMT{i}' for i in range(1, 7)]
columns_of_interest = pay_amt_columns + bill_amt_columns

# Function to detect outliers using the IQR method and replace them with NaN
def replace_outliers_with_nan(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        
        # Identify outliers
        outliers = df[(df[column] < Q1 - outlier_step) | (df[column] > Q3 + outlier_step)].index
        
        # Replace outliers with NaN
        df.loc[outliers, column] = np.nan
    
    return df

# Function to process a DataFrame: replace outliers and apply imputer
def process_dataframe(df, columns):
    # Replace outliers with NaN
    df = replace_outliers_with_nan(df, columns)
    
    # Initialize the IterativeImputer
    imputer = IterativeImputer(max_iter=10, random_state=0)
    
    # Apply the imputer to the relevant columns
    df[columns] = imputer.fit_transform(df[columns])
    
    return df

# Apply the process to cc_data
cc_data = process_dataframe(cc_data, columns_of_interest)

# Apply the process to pilot_data
pilot_data = process_dataframe(pilot_data, columns_of_interest)

# Verify the changes
print("cc_data head:")
print(cc_data.head())

print("\npilot_data head:")
print(pilot_data.head())
```

####  Now with the mean, skewness and kurtosis, we can implement the transformation as below

```python
import numpy as np
from scipy.stats import norm

def cornish_fisher_expansion(z_scores, skewness, kurtosis):
    """Adjust Z-scores using the Cornish-Fisher expansion."""
    z_adj = (z_scores + 
             (1/6) * (z_scores**2 - 1) * skewness + 
             (1/24) * (z_scores**3 - 3 * z_scores) * (kurtosis - 3) - 
             (1/36) * (2 * z_scores**3 - 5 * z_scores) * (skewness**2))
    return z_adj

# List of payment columns
payment_columns = ['PAY_1']

# Function to compute percentiles and adjusted z-scores for a dataframe
def compute_percentiles_and_z_scores(df):
    for column in payment_columns:
        pay_data = df[column]
        
        # Calculate statistics for the payment column
        mean = pay_data.mean()
        std = pay_data.std()
        skewness = pay_data.skew()
        kurtosis_value = pay_data.kurtosis()

        # Standardize the data
        z_scores = (pay_data - mean) / std

        # Adjust Z-scores using Cornish-Fisher expansion
        adjusted_z_scores = cornish_fisher_expansion(z_scores, skewness, kurtosis_value)

        # Convert adjusted Z-scores to percentiles
        percentiles = norm.cdf(adjusted_z_scores)

        # Create new columns in the dataframe for the percentiles and adjusted z-scores
        percentile_col_name = f'{column}_trans'
        z_score_col_name = f'{column}_z'

        df[percentile_col_name] = percentiles
        df[z_score_col_name] = adjusted_z_scores

# Compute percentiles and z-scores for cc_data_new
compute_percentiles_and_z_scores(cc_data_new)

# Compute percentiles and z-scores for pilot_data
compute_percentiles_and_z_scores(pilot_data)

# Now let's see the head of the updated dataframes
print("Head of cc_data_new:")
print(cc_data_new.head())

print("\nHead of pilot_data:")
print(pilot_data.head())
```

```python
import pandas as pd
from itertools import groupby  # Ensure this import is correctly included

# Function to create new payment behavior features
def create_payment_features(df):
    payment_columns = ['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

    # Average Payment Status
    df['avg_payment_status'] = df[payment_columns].mean(axis=1)

    # Maximum Payment Delay
    df['max_payment_delay'] = df[payment_columns].max(axis=1)

    # Consecutive Delays
    df['max_consecutive_delays'] = df[payment_columns].apply(
        lambda x: max((len(list(g)) for k, g in groupby(x > 0) if k), default=0), axis=1
    )



# Assuming cc_data_new, cc_data_test, and pilot_data are defined and available
create_payment_features(cc_data_new)
create_payment_features(pilot_data)
```

### How to interprete:  
AUC: If the curve is close to top left corner, it indicates the model provides a good sensitivity, in other words, the True Positive increase steadily with slow increase in False Positive. with the ROC curve and AUC, we can then decide which model provide better performance regarding to sensitivity. and by comparison, we can see the best model is the gbc with AUC = 0.78, though other two are some where around 0.77 and 0.76.   
  
Confusion curve: Have better understanding of the sensitivity and specificity, for later profit consideration, we should aim for higher TP and lower FP, with GBC, we have 35% of TP and 5% of FP.
  
Calibration curve: which is another plot that measure our prediction performance, it compares the predicted probability (x axis) and the emperical probability (y axis), simply speaking, the dotted line means that our prediction is exactly the same as the emprical, hence we would like our prediction much closer to the dotted line. here we can see the GBC performs well with less volatility, though it deviates as the probability getting bigger, ada boosting shows inconsistency between predicted and emprical (over predicting the default rate) , hence we will not consider ada as a model to use due to the unexpected performance in calibration. the deviation of gbc at the higher probability actually means that we are over predicting the default rate, which in a sense it might be good for the bank if the strategy is conservartive

```python
import numpy as np
from scipy import stats

# Define the thresholds and approval rates
thresholds = [0.2, 0.4, 0.5, 0.6, 0.8]
approval_rates_male = [0.56, 0.79, 0.83, 0.88, 0.99]
approval_rates_female = [0.66, 0.85, 0.90, 0.91, 0.99]

# Count occurrences of males and females in the 'SEX' column
male_count = cc_data_new[cc_data_new['SEX'] == 1]['SEX'].count()
female_count = cc_data_new[cc_data_new['SEX'] == 2]['SEX'].count()

# Perform two-sample t-tests for each threshold
alpha = 0.01  # Set the significance level to 0.01
results = []

for i in range(len(thresholds)):
    # Use approval rates and sample sizes as proportions
    p1 = approval_rates_male[i]
    p2 = approval_rates_female[i]
    n1 = male_count
    n2 = female_count
    
    # Calculate the t-statistic using the formula for comparing proportions
    t_statistic = (p1 - p2) / np.sqrt((p1 * (1 - p1)) / n1 + (p2 * (1 - p2)) / n2)
    
    # Calculate degrees of freedom using Welch's approximation
    df = ((p1 * (1 - p1) / n1) + (p2 * (1 - p2) / n2))**2 / (((p1 * (1 - p1)) / n1)**2 / (n1 - 1) + ((p2 * (1 - p2)) / n2)**2 / (n2 - 1))

    # Calculate p-value
    p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # Two-tailed test
    
    # Determine if the result is significant
    is_significant = p_value < alpha
    
    results.append((thresholds[i], t_statistic, p_value, is_significant))

# Print the results
print("Threshold | t-statistic | p-value | Significant (alpha = 0.01)")
for threshold, t_statistic, p_value, is_significant in results:
    print(f"{threshold:<9} | {t_statistic:.4f} | {p_value:.4f} | {is_significant}")

# Draw conclusions
conclusions = []

for threshold, t_statistic, p_value, is_significant in results:
    if is_significant:
        conclusion = f"At threshold {threshold}, there is a significant difference between male and female approval rates (t-statistic = {t_statistic:.4f}, p-value = {p_value:.4f})."
    else:
        conclusion = f"At threshold {threshold}, there is no significant difference between male and female approval rates (t-statistic = {t_statistic:.4f}, p-value = {p_value:.4f})."
    conclusions.append(conclusion)

# Print the conclusions
for conclusion in conclusions:
    print(conclusion)
```
